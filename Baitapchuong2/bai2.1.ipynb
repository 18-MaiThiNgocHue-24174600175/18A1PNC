```python
from xml.dom.minidom import Document
doc = Document()
company = doc.createElement("company")
doc.appendChild(company)
director = doc.createElement("director")
director.appendChild(doc.createTextNode("Nguyễn Văn A"))
company.appendChild(director)
address = doc.createElement("address")
address.appendChild(doc.createTextNode("123 Đường ABC, Phường 1, TP.HaNoi "))
company.appendChild(address)
phone = doc.createElement("phone")
phone.appendChild(doc.createTextNode("0123456789"))
company.appendChild(phone)
taxcode = doc.createElement("taxcode")
taxcode.appendChild(doc.createTextNode("0312345678"))
company.appendChild(taxcode)
with open("company.xml", "w", encoding="utf-8") as f:
    f.write(doc.toprettyxml(indent="    ", encoding="utf-8").decode("utf-8"))

print(" Đã tạo file company.xml thành công.")
```
