```python
import json
ds_nhan_vien = [
    {"ma_nv": "NV01", "ho_ten": "Nguyễn Văn A", "gioi_tinh": "Nam", "nam_sinh": 1990, "don_vi": "A1"},
    {"ma_nv": "NV02", "ho_ten": "Trần Thị B", "gioi_tinh": "Nữ", "nam_sinh": 1992, "don_vi": "A2"},
    {"ma_nv": "NV03", "ho_ten": "Lê Văn C", "gioi_tinh": "Nam", "nam_sinh": 1991, "don_vi": "A1"},
    {"ma_nv": "NV04", "ho_ten": "Phạm Thị D", "gioi_tinh": "Nữ", "nam_sinh": 1993, "don_vi": "A3"},
    {"ma_nv": "NV05", "ho_ten": "Nguyễn Văn E", "gioi_tinh": "Nam", "nam_sinh": 1995, "don_vi": "A2"}
]
with open("nhanvien.json", "w", encoding="utf-8") as f:
    json.dump(ds_nhan_vien, f, ensure_ascii=False, indent=4)

print(" Đã ghi dữ liệu nhân viên vào file nhanvien.json")
with open("nhanvien.json", "r", encoding="utf-8") as f:
    nhan_vien = json.load(f)
thong_ke = {}
for nv in nhan_vien:
    dv = nv["don_vi"]
    thong_ke[dv] = thong_ke.get(dv, 0) + 1

tong_nv = len(nhan_vien)
print("\nTên công ty: Công ty TNHH Đất Việt")
print("Địa chỉ: abc Giải Phóng – Hà Nội")
print("----- Thống kê nhân viên theo đơn vị -----")
stt = 1
for dv, so_nv in thong_ke.items():
    ty_le = (so_nv / tong_nv) * 100
    print(f"{stt}. Tên đơn vị: Đơn vị {dv}")
    print(f"   - Số nhân viên: {so_nv}")
    print(f"   - Tỷ lệ: {ty_le:.2f}%")
    stt += 1
```
