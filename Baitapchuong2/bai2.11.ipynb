```python
import json
from datetime import datetime
giao_dich = []
print("=== CHƯƠNG TRÌNH GHI GIAO DỊCH TIỀN & VÀNG ===")
print("Nhập thông tin giao dịch (để trống loại giao dịch để kết thúc)")
while True:
    loai = input("Loại giao dịch (Tiền/Vàng, Enter để thoát): ").strip()
    if loai == "":
        break
    so_tien = input("Số tiền/gia trị giao dịch: ").strip()
    mo_ta = input("Mô tả giao dịch: ").strip()
    giao_dich.append({
        "loai": loai,
        "so_tien": so_tien,
        "mo_ta": mo_ta
    })
if len(giao_dich) == 0:
    print("Không có giao dịch nào để ghi.")
else:
    lua_chon = input("Bạn có muốn ghi vào file không? (1 = Có, 0 = Không): ").strip()
    if lua_chon == "1":
        thoi_gian = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
        ten_file = f"{thoi_gian}.json"
        with open(ten_file, "w", encoding="utf-8") as f:
            json.dump(giao_dich, f, ensure_ascii=False, indent=4)

        print(f" Đã ghi {len(giao_dich)} giao dịch vào file {ten_file}")
    else:
        print("Bạn đã chọn không ghi dữ liệu.")
```
