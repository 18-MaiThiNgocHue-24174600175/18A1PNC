```python
import requests
url = "https://jsonplaceholder.typicode.com/comments?postId=1"
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    tong_so_bai = len(data)
    print(f"✅ Tổng số bài post nổi bật (comment): {tong_so_bai}")

    print("\n===== DANH SÁCH CÁC BÀI POST NỔI BẬT =====")
    for i, post in enumerate(data[:3], start=1):
        print(f"\nBài số {i}:")
        print(f"Post ID: {post['postId']}")
        print(f"ID: {post['id']}")
        print(f"Name: {post['name']}")
        print(f"Email: {post['email']}")
        print(f"Body: {post['body']}")
        print("-" * 40)

else:
    print(f"Lỗi khi truy cập API. Mã lỗi: {response.status_code}")
```
