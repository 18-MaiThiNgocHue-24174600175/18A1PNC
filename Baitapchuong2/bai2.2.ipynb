```python
from xml.dom.minidom import Document
doc = Document()
root = doc.createElement("students")
doc.appendChild(root)
students = [
    {"id": "SV001", "name": "Mai Văn A", "year": "2003", "class": "KTPM1", "gender": "Nam"},
    {"id": "SV002", "name": "Trần Thị B", "year": "2002", "class": "KTPM2", "gender": "Nữ"},
    {"id": "SV003", "name": "Nguyễn Văn C", "year": "2003", "class": "KTPM3", "gender": "Nam"}
]
for sv in students:
    student = doc.createElement("student")

    id_tag = doc.createElement("id")
    id_tag.appendChild(doc.createTextNode(sv["id"]))
    student.appendChild(id_tag)

    name_tag = doc.createElement("name")
    name_tag.appendChild(doc.createTextNode(sv["name"]))
    student.appendChild(name_tag)

    year_tag = doc.createElement("year")
    year_tag.appendChild(doc.createTextNode(sv["year"]))
    student.appendChild(year_tag)

    class_tag = doc.createElement("class")
    class_tag.appendChild(doc.createTextNode(sv["class"]))
    student.appendChild(class_tag)

    gender_tag = doc.createElement("gender")
    gender_tag.appendChild(doc.createTextNode(sv["gender"]))
    student.appendChild(gender_tag)

    root.appendChild(student)
with open("students.xml", "w", encoding="utf-8") as f:
    f.write(doc.toprettyxml(indent="    ", encoding="utf-8").decode("utf-8"))

print("Đã tạo file students.xml thành công.")
```
