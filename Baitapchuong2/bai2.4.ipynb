```python
from xml.dom import minidom
doc = minidom.parse("sample.xml")
company = doc.documentElement
print("Tên node gốc:", company.tagName)
name = company.getElementsByTagName("name")[0].firstChild.data
print("Tên công ty:", name)
staffs = company.getElementsByTagName("staff")
for s in staffs:
    staff_id = s.getAttribute("id")
    staff_name = s.getElementsByTagName("name")[0].firstChild.data
    staff_salary = s.getElementsByTagName("salary")[0].firstChild.data
    print(f" Nhân viên {staff_id}: {staff_name}, Lương: {staff_salary}")
```
