```python
import requests
from xml.dom import minidom
import csv
url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"
response = requests.get(url)
with open("rss.xml", "wb") as f:
    f.write(response.content)
doc = minidom.parse("rss.xml")
items = doc.getElementsByTagName("item")
with open("rss_news.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Tiêu đề", "Link", "Ngày đăng"])  
    for item in items:
        title = item.getElementsByTagName("title")[0].firstChild.data
        link = item.getElementsByTagName("link")[0].firstChild.data
        pubDate = item.getElementsByTagName("pubDate")[0].firstChild.data
        writer.writerow([title, link, pubDate])
print("Đã lưu tin tức RSS vào file rss_news.csv")
```
