```python
import json
person = {
    "name": "Nguyễn Văn A ",
    "age": 21,
    "city": "Hà Nội"
}
json_string = json.dumps(person, indent=4, ensure_ascii=False)
print(json_string)
with open("person.json", "w", encoding="utf-8") as f:
    f.write(json_string)

print(" Đã lưu dữ liệu vào person.json")
```
