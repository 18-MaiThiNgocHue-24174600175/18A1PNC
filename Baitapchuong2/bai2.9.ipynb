```python
import json
staff = {
    "phone": "0123456789",
    "name": "Nguyễn Văn A ",
    "address": "Hà Nội",
    "id": "NV001"
}
json_sorted = json.dumps(staff, indent=4, sort_keys=True, ensure_ascii=False)
print(json_sorted)
with open("staff.json", "w", encoding="utf-8") as f:
    f.write(json_sorted)
```
