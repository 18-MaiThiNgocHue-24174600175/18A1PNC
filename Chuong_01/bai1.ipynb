```python
class HinhChuNhat:
    def __init__(self, dai=0, rong=0):
        self.chieu_dai = dai
        self.chieu_rong = rong

    def nhap_du_lieu(self):
        print("--- Nhập thông tin Hình chữ nhật ---")
        while True:
            try:
                self.chieu_dai = float(input("Nhập chiều dài: "))
                if self.chieu_dai <= 0:
                    raise ValueError
                break
            except ValueError:
                print("Chiều dài phải là số dương. Vui lòng nhập lại.")
        
        while True:
            try:
                self.chieu_rong = float(input("Nhập chiều rộng: "))
                if self.chieu_rong <= 0:
                    raise ValueError
                break
            except ValueError:
                print("Chiều rộng phải là số dương. Vui lòng nhập lại.")

    def tinh_chu_vi(self):
        """Tính và trả về Chu vi."""
        return 2 * (self.chieu_dai + self.chieu_rong)

    def tinh_dien_tich(self):
        """Tính và trả về Diện tích."""
        return self.chieu_dai * self.chieu_rong

    def in_thong_tin(self):
        """In tất cả thông tin của Hình chữ nhật ra màn hình."""
        chu_vi = self.tinh_chu_vi()
        dien_tich = self.tinh_dien_tich()
        print("\n--- Thông tin Hình chữ nhật ---")
        print(f"Chiều dài: {self.chieu_dai}")
        print(f"Chiều rộng: {self.chieu_rong}")
        print(f"Chu vi: {chu_vi}")
        print(f"Diện tích: {dien_tich}")

# --- Chương trình chính ---
print("Bài tập 1: Hình chữ nhật")
hcn1 = HinhChuNhat() 
hcn1.nhap_du_lieu()
hcn1.in_thong_tin()
```
