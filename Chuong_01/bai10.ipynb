```python
import math

# Lớp Điểm
class Diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def hien_thi(self):
        print(f"Tọa độ điểm: ({self.x}, {self.y})")

# Lớp Elip kế thừa từ Điểm
class Elip(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)   # gọi lại hàm khởi tạo lớp cha
        self.a = a
        self.b = b

    def dien_tich(self):
        return math.pi * self.a * self.b

    def hien_thi(self):
        print(f"Tâm elip: ({self.x}, {self.y})")
        print(f"Bán trục lớn a = {self.a}, bán trục nhỏ b = {self.b}")

# Lớp Đường tròn kế thừa từ Elip
class DuongTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
        self.r = r

    def dien_tich(self):
        return math.pi * self.r ** 2

    def hien_thi(self):
        print(f"Tâm đường tròn: ({self.x}, {self.y})")
        print(f"Bán kính r = {self.r}")

# ---------------------------
# Chạy thử chương trình
if __name__ == "__main__":
    print("=== TÍNH DIỆN TÍCH ELIP VÀ ĐƯỜNG TRÒN ===")

    # Nhập thông tin elip
    x = float(input("Nhập hoành độ tâm elip: "))
    y = float(input("Nhập tung độ tâm elip: "))
    a = float(input("Nhập bán trục lớn a: "))
    b = float(input("Nhập bán trục nhỏ b: "))

    elip = Elip(x, y, a, b)
    elip.hien_thi()
    print(f"Diện tích elip: {elip.dien_tich():.2f}")

    # Nhập thông tin đường tròn
    x = float(input("\nNhập hoành độ tâm đường tròn: "))
    y = float(input("Nhập tung độ tâm đường tròn: "))
    r = float(input("Nhập bán kính r: "))

    dt = DuongTron(x, y, r)
    dt.hien_thi()
    print(f"Diện tích đường tròn: {dt.dien_tich():.2f}")
```
