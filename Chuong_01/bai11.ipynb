```python
import math

# -------------------------------
# Lớp Tam giác
class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        p = self.chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

    def hien_thi(self):
        print(f"Cạnh a = {self.a}, b = {self.b}, c = {self.c}")
        print(f"Chu vi: {self.chu_vi():.2f}")
        print(f"Diện tích: {self.dien_tich():.2f}")

# -------------------------------
# Lớp Tam giác vuông kế thừa từ Tam giác
class TamGiacVuong(TamGiac):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
        # sắp xếp để c là cạnh lớn nhất (cạnh huyền)
        sides = sorted([a, b, c])
        self.a, self.b, self.c = sides[0], sides[1], sides[2]

    def la_tam_giac_vuong(self):
        return math.isclose(self.a**2 + self.b**2, self.c**2, rel_tol=1e-9)

    def hien_thi(self):
        super().hien_thi()
        if self.la_tam_giac_vuong():
            print("Đây là tam giác vuông.")
        else:
            print("Không phải tam giác vuông.")

# -------------------------------
# Lớp Tam giác cân kế thừa từ Tam giác
class TamGiacCan(TamGiac):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)

    def la_tam_giac_can(self):
        return self.a == self.b or self.b == self.c or self.a == self.c

    def hien_thi(self):
        super().hien_thi()
        if self.la_tam_giac_can():
            print("Đây là tam giác cân.")
        else:
            print("Không phải tam giác cân.")

# -------------------------------
# Lớp Tam giác đều kế thừa từ Tam giác cân
class TamGiacDeu(TamGiacCan):
    def __init__(self, a):
        super().__init__(a, a, a)  # tam giác đều có 3 cạnh bằng nhau

    def hien_thi(self):
        super().hien_thi()
        print("Đây là tam giác đều.")

# -------------------------------
# Chạy thử chương trình
if __name__ == "__main__":
    print("=== TÍNH TOÁN CÁC LOẠI TAM GIÁC ===")

    # Tam giác thường
    a = float(input("Nhập cạnh a của tam giác: "))
    b = float(input("Nhập cạnh b của tam giác: "))
    c = float(input("Nhập cạnh c của tam giác: "))
    tg = TamGiac(a, b, c)
    tg.hien_thi()

    # Tam giác vuông
    print("\n=== TAM GIÁC VUÔNG ===")
    a = float(input("Nhập cạnh a: "))
    b = float(input("Nhập cạnh b: "))
    c = float(input("Nhập cạnh c: "))
    tgv = TamGiacVuong(a, b, c)
    tgv.hien_thi()

    # Tam giác cân
    print("\n=== TAM GIÁC CÂN ===")
    a = float(input("Nhập cạnh a: "))
    b = float(input("Nhập cạnh b: "))
    c = float(input("Nhập cạnh c: "))
    tgc = TamGiacCan(a, b, c)
    tgc.hien_thi()

    # Tam giác đều
    print("\n=== TAM GIÁC ĐỀU ===")
    a = float(input("Nhập cạnh a của tam giác đều: "))
    tgd = TamGiacDeu(a)
    tgd.hien_thi()
```
