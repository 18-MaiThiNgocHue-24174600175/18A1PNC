```python
class ThiSinh:
    def __init__(self, ho_ten="", toan=0, ly=0, hoa=0):
        self.ho_ten = ho_ten
        self.diem_toan = toan
        self.diem_ly = ly
        self.diem_hoa = hoa

    def nhap_thong_tin(self):
        self.ho_ten = input("Nhập Họ tên thí sinh: ")
        def nhap_diem(ten_mon):
            while True:
                try:
                    diem = float(input(f"Nhập điểm {ten_mon} (0-10): "))
                    if 0 <= diem <= 10:
                        return diem
                    else:
                        print("Điểm phải nằm trong khoảng từ 0 đến 10. Vui lòng nhập lại.")
                except ValueError:
                    print("Nhập không hợp lệ. Vui lòng nhập một số.")
        
        self.diem_toan = nhap_diem("Toán")
        self.diem_ly = nhap_diem("Lý")
        self.diem_hoa = nhap_diem("Hoá")

    def tinh_tong_diem(self):
        return self.diem_toan + self.diem_ly + self.diem_hoa

    def in_thong_tin(self):
        tong_diem = self.tinh_tong_diem()
        print(f"| {self.ho_ten:<20} | {self.diem_toan:<5.1f} | {self.diem_ly:<5.1f} | {self.diem_hoa:<5.1f} | {tong_diem:<9.1f} |")
def quan_ly_thi_sinh():
    print("\nBài 2: Quản lý Thí sinh")
    danh_sach_thi_sinh = []
    so_luong = 0
    while so_luong <= 0:
        try:
            so_luong = int(input("Nhập số lượng thí sinh cần thêm: "))
            if so_luong <= 0:
                print("Số lượng phải lớn hơn 0.")
        except ValueError:
            print("Vui lòng nhập một số nguyên.")
            
    for i in range(so_luong):
        print(f"\n--- Nhập thông tin thí sinh thứ {i+1} ---")
        ts = ThiSinh()
        ts.nhap_thong_tin()
        danh_sach_thi_sinh.append(ts)
    DIEM_CHUAN = 0
    while DIEM_CHUAN <= 0:
        try:
            DIEM_CHUAN = float(input("\nNhập điểm chuẩn vào trường: "))
            if DIEM_CHUAN <= 0:
                print("Điểm chuẩn phải là số dương.")
        except ValueError:
            print("Vui lòng nhập một số.")
    print("\n" + "="*70)
    print("DANH SÁCH THÍ SINH TRÚNG TUYỂN")
    print("="*70)
    print("| Họ tên | Toán  | Lý    | Hoá   | Tổng điểm |")
    print("-" * 70)
    
    co_thi_sinh_trung_tuyen = False
    for ts in danh_sach_thi_sinh:
        if ts.tinh_tong_diem() >= DIEM_CHUAN:
            ts.in_thong_tin()
            co_thi_sinh_trung_tuyen = True

    if not co_thi_sinh_trung_tuyen:
        print("|Không có thí sinh nào trúng tuyển.|")
    print("-" * 70)
quan_ly_thi_sinh()
    
```
