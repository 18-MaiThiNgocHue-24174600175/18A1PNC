```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.items = [0] * capacity
        self.top = -1

    def push(self, value):
        if self.isFull():
            raise OverflowError("Ngăn xếp đã đầy!")
        self.top += 1
        self.items[self.top] = value

    def pop(self):
        if self.isEmpty():
            raise IndexError("Ngăn xếp rỗng!")
        value = self.items[self.top]
        self.top -= 1
        return value

    def count(self):
        return self.top + 1

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.capacity - 1
```
