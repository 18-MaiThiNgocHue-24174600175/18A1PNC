```python
class Date:
    def __init__(self, day=1, month=1, year=2023):
        self.day = day
        self.month = month
        self.year = year
    def display(self):
        print(f"{self.day}/{self.month}/{self.year}")
    def next(self):
        def is_leap(year):
            return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
        days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if is_leap(self.year):
            days_in_month[2] = 29
        max_day = days_in_month[self.month]
        self.day += 1
        if self.day > max_day:
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1
        return self
    

```
