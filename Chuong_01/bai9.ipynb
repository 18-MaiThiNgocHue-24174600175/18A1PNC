```python
import math

# Lớp đa giác
class DaGiac:
    def __init__(self, so_canh, danh_sach_canh):
        self.so_canh = so_canh
        self.danh_sach_canh = danh_sach_canh

    def chu_vi(self):
        return sum(self.danh_sach_canh)

# Lớp hình bình hành kế thừa từ đa giác
class HinhBinhHanh(DaGiac):
    def __init__(self, day, canh_ben, chieu_cao):
        super().__init__(4, [day, canh_ben, day, canh_ben])
        self.day = day
        self.canh_ben = canh_ben
        self.chieu_cao = chieu_cao

    def dien_tich(self):
        return self.day * self.chieu_cao

# Lớp hình chữ nhật kế từ hình bình hành
class HinhChuNhat(HinhBinhHanh):
    def __init__(self, dai, rong):
        super().__init__(dai, rong, rong)  # chiều cao bằng chiều rộng
        self.dai = dai
        self.rong = rong

    def dien_tich(self):
        return self.dai * self.rong

# Lớp hình vuông kế từ hình chữ nhật
class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        super().__init__(canh, canh)
        self.canh = canh

    def dien_tich(self):
        return self.canh ** 2

# ---------------------------
# Chạy thử chương trình
if __name__ == "__main__":
    print("=== TÍNH CHU VI & DIỆN TÍCH CÁC HÌNH ===")

    # Hình bình hành
    day = float(input("Nhập đáy hình bình hành: "))
    canh_ben = float(input("Nhập cạnh bên hình bình hành: "))
    chieu_cao = float(input("Nhập chiều cao hình bình hành: "))
    hbh = HinhBinhHanh(day, canh_ben, chieu_cao)
    print("Chu vi HBH:", hbh.chu_vi())
    print("Diện tích HBH:", hbh.dien_tich())

    # Hình chữ nhật
    dai = float(input("Nhập chiều dài hình chữ nhật: "))
    rong = float(input("Nhập chiều rộng hình chữ nhật: "))
    hcn = HinhChuNhat(dai, rong)
    print("Chu vi HCN:", hcn.chu_vi())
    print("Diện tích HCN:", hcn.dien_tich())

    # Hình vuông
    canh = float(input("Nhập cạnh hình vuông: "))
    hv = HinhVuong(canh)
    print("Chu vi HV:", hv.chu_vi())
    print("Diện tích HV:", hv.dien_tich())
```
