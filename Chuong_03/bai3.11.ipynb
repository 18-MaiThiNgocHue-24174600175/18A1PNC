```python
import pandas as pd
import numpy as np
import io
data_s1 = """
date,symbol,open,high,low,close,volume
2023-10-01,ABC,10.0,10.5,9.8,10.3,100000
2023-10-02,ABC,10.3,10.8,10.2,10.7,120000
2023-10-01,XYZ,50.0,51.0,49.5,50.5,50000
2023-10-02,XYZ,50.5,52.0,50.0,51.5,60000
2023-10-03,ABC,10.7,11.0,10.5,10.9,110000
2023-10-03,XYZ,51.5,53.0,51.0,52.5,70000
"""
data_s2 = """
date,symbol,open,high,low,close,volume
2023-10-04,ABC,10.9,11.5,10.8,11.2,130000
2023-10-04,XYZ,52.5,,52.0,53.0,80000
2023-10-05,ABC,11.2,12.0,11.0,11.8,150000
2023-10-05,XYZ,53.0,54.0,,53.5,90000
2023-10-06,ABC,11.8,12.2,11.5,12.1,140000
2023-10-06,XYZ,53.5,55.0,53.0,54.5,100000
"""
data_comp = """
symbol,name,employees,headquarters_city,headquarters_state
ABC,Alpha Beta Corp,500,New York,NY
XYZ,Xylos Zenith Inc,1200,San Francisco,CA
"""
stocks1 = pd.read_csv(io.StringIO(data_s1))
stocks2 = pd.read_csv(io.StringIO(data_s2))
companies = pd.read_csv(io.StringIO(data_comp))

```

```python
print("--- 1.a) Thông tin stocks1 ---")
print("5 dòng đầu (Head):")
print(stocks1.head())
print("\n5 dòng cuối (Tail):")
print(stocks1.tail())
print("\nKiểu dữ liệu (Dtype):")
print(stocks1.dtypes)
print("\nThông tin (Info):")
stocks1.info()
print("-" * 50)
print("--- 1.b) Thông tin stocks2 ---")
print("5 dòng đầu (Head):")
print(stocks2.head())
print("\n5 dòng cuối (Tail):")
print(stocks2.tail())
print("\nKiểu dữ liệu (Dtype):")
print(stocks2.dtypes)
print("\nThông tin (Info):")
stocks2.info()
print("-" * 50)
print("--- 1.c) Thông tin companies ---")
print("Dữ liệu của companies:")
print(companies)
print("\nKiểu dữ liệu (Dtype):")
print(companies.dtypes)
print("\nThông tin (Info):")
companies.info()
print("-" * 50)
```

```python
print("--- 2. Xử lý giá trị Null trong stocks2 ---")
print("Trước khi xử lý Null:")
print(stocks2[stocks2.isnull().any(axis=1)])
stocks2['high'] = stocks2.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks2['low'] = stocks2.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))
print("\nSau khi xử lý Null:")
print(stocks2.loc[3:4]) 
print("-" * 50)
```

```python
stocks = pd.concat([stocks1, stocks2], axis=0, ignore_index=True)

print("--- 3. DataFrame stocks sau khi gộp (Concatenate) ---")
print("15 dòng cuối (Tail) của stocks:")
print(stocks.tail(15)) 
print("-" * 50)
stocks_companies = pd.merge(stocks, companies, on='symbol', how='left')

print("--- 4. DataFrame stocks_companies sau khi gộp (Merge) ---")
print("5 dòng đầu (Head) của stocks_companies:")
print(stocks_companies.head())
print("-" * 50)
```

```python
avg_metrics_by_company = stocks_companies.groupby('name')[['open', 'high', 'low', 'close', 'volume']].mean().round(2)

print("--- 5. Giá trị trung bình của các chỉ số theo công ty ---")
print(avg_metrics_by_company)
print("-" * 50)
```

```python
close_metrics_by_company = stocks_companies.groupby('name')['close'].agg(['mean', 'max', 'min']).round(2)
print("--- 6. Giá đóng cửa (close) trung bình, lớn nhất và nhỏ nhất ở mỗi công ty ---")
print(close_metrics_by_company)
print("-" * 50)
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
```

```python
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])

print("--- 7. Tạo cột parsed_time và kiểu dữ liệu ---")
print("5 dòng đầu của cột parsed_time:")
print(stocks_companies['parsed_time'].head())
print(f"\nKiểu dữ liệu của cột parsed_time: **{stocks_companies['parsed_time'].dtype}**")
print("-" * 50)
```

```python
stocks_companies['result'] = np.where(stocks_companies['close'] > stocks_companies['open'], 'up', 'down')

print("--- 8. Thêm cột result ---")
print(stocks_companies[['date', 'symbol', 'open', 'close', 'result']])
print("-" * 50)
```
