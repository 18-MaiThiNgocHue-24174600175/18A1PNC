```python
baseball_2D = [
    [74, 180], [74, 215], [72, 210], [73, 210], [75, 188], [76, 176], [74, 209],
    [75, 200], [74, 231], [72, 180], [73, 188], [73, 180], [74, 185], [74, 188],
    [75, 200], [73, 180], [74, 215], [74, 210], [75, 200], [74, 215], [74, 200],
    [73, 190], [75, 200], [75, 185], [74, 195], [74, 200], [73, 205], [74, 192],
    [72, 190], [73, 170],[75, 195], [76, 210], [74, 205], [73, 190], [75, 220],
    [76, 230], [74, 185], [75, 205], [73, 195], [72, 185],
    [74, 210], [75, 215], [73, 175], [74, 190], [76, 225],
    [75, 200], [74, 190], [73, 175], [75, 205], [74, 200] 
]

import numpy as np
np_baseball = np.array(baseball_2D)

print("1. Thông tin về np_baseball:")
print(f"Kiểu dữ liệu: {np_baseball.dtype}")
print(f"Kích thước (Shape): {np_baseball.shape}")
print(f"Mảng:\n{np_baseball}")
row_50  = np_baseball[49]

print("\n2. Giá trị của dòng thứ 50:")
print(f"Chiều cao (inch) và cân nặng (pounds): {row_50 }")
np_weight = np_baseball[:, 1]

print("\n3. Mảng np_weight (Cân nặng pounds):")
print(np_weight)
player_index = 12
if player_index < np_baseball.shape[0]:
    height_of_player = np_baseball[player_index, 0]
    print(f"\n4. Chiều cao của vận động viên thứ {player_index + 1} (index {player_index}): {height_of_player} inches")
else:
    print(f"\n4. KHÔNG tìm thấy vận động viên thứ 124 trong dữ liệu hiện tại (chỉ có {np_baseball.shape[0]} người).")
    print("   **Cách truy cập (nếu có đủ dữ liệu):** np_baseball[123, 0]")
mean_height = np.mean(np_baseball[:, 0])
mean_weight = np.mean(np_baseball[:, 1])
print("\n5. Thống kê:")
print(f"- Chiều cao trung bình: {mean_height:.2f} inches")
print(f"- Cân nặng trung bình: {mean_weight:.2f} pounds")
heights = np_baseball[:, 0]
weights = np_baseball[:, 1]
correlation_matrix = np.corrcoef(heights, weights)
correlation_coefficient = correlation_matrix[0, 1]
print("\n6. Nhận xét Mối tương quan:")
print(f"Hệ số tương quan giữa chiều cao và cân nặng: {correlation_coefficient:.4f}")
if correlation_coefficient > 0.5:
    print("-> **Tương quan thuận mạnh (Strong Positive Correlation).** Khi chiều cao tăng, cân nặng có xu hướng tăng theo.")
elif correlation_coefficient > 0:
    print("-> **Tương quan thuận (Positive Correlation).** Khi chiều cao tăng, cân nặng có xu hướng tăng nhẹ.")
elif correlation_coefficient < -0.5:
    print("-> **Tương quan nghịch mạnh (Strong Negative Correlation).**")
else:
    print("-> **Không có tương quan hoặc tương quan rất yếu.**")
```
