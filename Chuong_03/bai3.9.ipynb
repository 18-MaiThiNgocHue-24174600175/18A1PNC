```python
import pandas as pd
import io
data = """
Team,Goals,Shots on target,Shots Off Target,Shooting Accuracy,Yellow Cards,Red Cards,Possession
Spain,12,42,33,55.9%,11,0,59%
Germany,10,32,32,47.8%,10,0,56.2%
Italy,6,34,45,43.0%,16,0,52.8%
Portugal,6,27,24,50.0%,12,0,49.2%
Russia,5,9,31,22.5%,10,0,56.0%
Greece,5,11,18,35.2%,11,1,47.2%
Croatia,4,13,12,51.9%,9,0,45.7%
England,5,17,29,41.9%,5,0,44.9%
France,3,22,24,37.9%,6,0,53.5%
Ukraine,2,10,13,47.1%,5,0,53.0%
Netherlands,2,20,20,33.1%,5,0,54.3%
Sweden,5,17,19,47.9%,7,0,43.1%
Poland,2,15,23,39.4%,7,1,52.0%
Denmark,4,10,10,50.0%,4,0,44.2%
"""

df_euro12 = pd.read_csv(io.StringIO(data))
df_euro12['Shooting Accuracy'] = df_euro12['Shooting Accuracy'].str.replace('%', '').astype(float) / 100

print("--- DataFrame ban đầu (df_euro12) ---")
print(df_euro12.head())
print("-" * 50)
goals_values = df_euro12['Goals']

print("1. Giá trị cột 'Goals':")
print(goals_values)
num_teams = df_euro12.shape[0]

print(f"2. Có **{num_teams}** đội tham gia Euro 2012.")
print("3. Thông tin chi tiết của DataFrame Euro 2012:")
df_euro12.info()
df_discipline = df_euro12[['Team', 'Yellow Cards', 'Red Cards']]

print("4. DataFrame mới chỉ chứa 'Team', 'Yellow Cards', 'Red Cards':")
print(df_discipline)
df_discipline_sorted = df_discipline.sort_values(
    by=['Red Cards', 'Yellow Cards'],
    ascending=False
)

print("5. Sắp xếp discipline giảm dần theo 'Red Cards' và 'Yellow Cards':")
print(df_discipline_sorted)
mean_yellow_cards = df_euro12['Yellow Cards'].mean()

print(f"6.a) Trung bình 'Yellow Cards' của tất cả các đội là: **{mean_yellow_cards:.2f}**")

teams_over_6_goals = df_euro12[df_euro12['Goals'] >= 6]

print("6.b) Các đội đã ghi hơn hoặc bằng 6 bàn thắng:")
print(teams_over_6_goals[['Team', 'Goals']])
teams_start_with_g = df_euro12[df_euro12['Team'].str.startswith('G')]

print("7. Các đội mà tên bắt đầu bằng 'G':")
print(teams_start_with_g[['Team']])
first_7_columns = df_euro12.iloc[:, :7]

print("8. 7 cột đầu tiên của euro12:")
print(first_7_columns.head())
all_except_last_3_columns = df_euro12.iloc[:, :-3]

print("9. Tất cả các cột, trừ 3 cột cuối:")
print(all_except_last_3_columns.head())
selected_columns = df_euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']]

print("10. Các cột 'Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards':")
print(selected_columns.head())
teams_to_show = ['England', 'Italy', 'Russia']
df_filtered_teams = df_euro12[df_euro12['Team'].isin(teams_to_show)]
result_11 = df_filtered_teams[['Team', 'Shooting Accuracy']]
print("11. Các cột 'Team', 'Shooting Accuracy' từ 'England', 'Italy', 'Russia':")
print(result_11)
```
