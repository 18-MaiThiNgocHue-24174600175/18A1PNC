```python
import numpy as np
heights = [185, 192, 178, 188, 175, 190, 180, 183, 195, 170]
positions = ['GK', 'A', 'M', 'D', 'M', 'A', 'GK', 'D', 'A', 'M']
np_positions = np.array(positions)

print(f"1.a) np_positions: {np_positions}")
print(f"Kiểu dữ liệu của np_positions: {np_positions.dtype}")
np_heights = np.array(heights)

print(f"1.b) np_heights: {np_heights}")
print(f"Kiểu dữ liệu của np_heights: {np_heights.dtype}")
is_gk = (np_positions == 'GK')
gk_heights = np_heights[is_gk]
average_gk_height = np.mean(gk_heights)

print(f"2. Chiều cao của các GK: {gk_heights}")
print(f"Chiều cao trung bình của các GK: {average_gk_height:.2f}")
not_gk = (np_positions != 'GK')
other_heights = np_heights[not_gk]
average_other_height = np.mean(other_heights)
print(f"3. Chiều cao của các vị trí khác (không phải GK): {other_heights}")
print(f"Chiều cao trung bình của những vị trí khác: {average_other_height:.2f}")
player_dtype = np.dtype([('position', 'U5'), ('height', 'float')])
player_data = np.array(list(zip(positions, heights)), dtype=player_dtype)
print("4. Mảng dữ liệu có cấu trúc 'player_data':")
print(player_data)
print(f"Kiểu dữ liệu của player_data: {player_data.dtype}")
sorted_players = np.sort(player_data, order='height')
print("\n5. Mảng players đã sắp xếp theo chiều cao (tăng dần):")
print(sorted_players)
lowest_player = sorted_players[0]
print(f"\nNgười chơi có chiều cao thấp nhất:")
print(f"- Chiều cao: {lowest_player['height']} (cm)")
print(f"- Vị trí: {lowest_player['position']}")
highest_player = sorted_players[-1]
print(f"\nNgười chơi có chiều cao cao nhất:")
print(f"- Chiều cao: {highest_player['height']} (cm)")
print(f"- Vị trí: {highest_player['position']}")
```
