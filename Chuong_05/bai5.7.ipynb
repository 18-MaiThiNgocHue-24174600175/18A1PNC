```python
import sqlite3
DB_FILE_5_7 = "bai_tap_5_5.db" 

print(f"\n--- BÀI 5.7: CẬP NHẬT TẤT CẢ GIÁ TRỊ ---")

conn = None
try:
    conn = sqlite3.connect(DB_FILE_5_7)
    cursor = conn.cursor()
    cursor.execute("SELECT Gia FROM SanPham WHERE ID = 1")
    gia_cu = cursor.fetchone()[0]

    cursor.execute("UPDATE SanPham SET Gia = Gia + 1000")
    conn.commit()
    
    print(f"✨ Đã cập nhật thành công {cursor.rowcount} bản ghi.")
    
    cursor.execute("SELECT Gia FROM SanPham WHERE ID = 1")
    gia_moi = cursor.fetchone()[0]
    
    print(f"   Giá cũ của SP ID=1: {gia_cu:.0f}")
    print(f"   Giá mới của SP ID=1: {gia_moi:.0f}")
    
except sqlite3.Error as e:
    print(f"❌ Lỗi xảy ra: {e}")
finally:
    if conn:
        conn.close()
```
