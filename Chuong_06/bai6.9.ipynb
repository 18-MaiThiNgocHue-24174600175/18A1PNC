```python
import threading
import time
from datetime import datetime

class WebCrawlerMock(threading.Thread):
    def __init__(self, name, actions):
        super().__init__(name=name)
        self.actions = actions 
        self.start_timestamp = None 

    def run(self):
    
        while self.start_timestamp is None:
            time.sleep(0.001)
        
        print(f"Starting {self.name}")
        
        for delay, message_type in self.actions:
            current_elapsed_time = time.time() - self.start_timestamp
            sleep_duration = delay - current_elapsed_time
            
            if sleep_duration > 0:
                time.sleep(sleep_duration)
            
            
            if message_type == "PRINT":
                
                mock_dt = datetime.fromtimestamp(self.start_timestamp + delay)
                
                current_time_str = mock_dt.strftime("Web %b %d %H:%M:%S %Y") 
                print(f"{self.name}: {current_time_str}")
            elif message_type == "EXIT":
                print(f"Exiting {self.name}")
                break 

def bai_6_9_thuc_thi_nhieu_luong():
    print("Starting Main Thread")
    
    # Google: In (0, 1, 2, 3), thoát (4)
    google_actions = [
        (0, "PRINT"),  # 10:25:52
        (1, "PRINT"),  # 10:25:53
        (2, "PRINT"),  # 10:25:54
        (3, "PRINT"),  # 10:25:55
        (4, "EXIT")    # Exiting Google
    ]

    # Yahoo: In (1, 3, 5, 7), thoát (8)
    yahoo_actions = [
        (1, "PRINT"),  # 10:25:53
        (3, "PRINT"),  # 10:25:55
        (5, "PRINT"),  # 10:25:57
        (7, "PRINT"),  # 10:25:59
        (8, "EXIT")    # Exiting Yahoo
    ]

    # Facebook: In (2, 5, 8, 11), thoát (12)
    facebook_actions = [
        (2, "PRINT"),  # 10:25:54
        (5, "PRINT"),  # 10:25:57
        (8, "PRINT"),  # 10:26:00
        (11, "PRINT"), # 10:26:03
        (12, "EXIT")   # Exiting Facebook
    ]

    threads = [
        WebCrawlerMock(name="Google", actions=google_actions),
        WebCrawlerMock(name="Yahoo", actions=yahoo_actions),
        WebCrawlerMock(name="Facebook", actions=facebook_actions)
    ]

    threads_start_actual_time = time.time() 

    for t in threads:
        t.start_timestamp = threads_start_actual_time 
        t.start()

    for t in threads:
        t.join()

    print("Exiting Main Thread")

bai_6_9_thuc_thi_nhieu_luong()
        
```
